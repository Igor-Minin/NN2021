```python
import tensorflow as tf
import numpy as np
import pandas as pd

npList = np.random.rand(4,5)
tfList = tf.constant(npList)
print("Numpy list:\n", npList)
print("Tensor list:\n", tfList)

print("Shape of tensor - ", tfList.shape)
print("Rank - ", tfList.ndim)
print("Size of tensor - ", tf.size(tfList).numpy())
axises = tfList.shape
counter = 0
for i in axises:
  print("Elements along axis", counter, "-", i)
  counter+=1

auxList = np.random.rand(4,5)
print("Auxiliary list for math operations:\n", auxList)

print("Addition scalar to tensor:\n", tfList+0.1)
print("Addition same dimensional list to tensor:\n", tfList+auxList)

print("Subtraction scalar from tensor:\n", tfList-0.1)
print("Subtraction same dimensional list from tensor:\n", tfList-auxList)

print("Division scalar on tensor:\n", tfList/2)
print("Division same dimensional list on tensor:\n", tfList/auxList)

print("Multiplication scalar on tensor:\n", tfList*2)
auxList = tf.reshape(auxList,shape = (5,4))
print("Multiplication same dimensional list on tensor:\n", tf.matmul(tfList,auxList))

d = {
    "one": pd.Series(np.arange(10, 20)),
    "two": pd.Series(np.arange(15, 30)),
    "three": pd.Series(np.arange(10,30))
}

df = pd.DataFrame(d)
print("Initial DataFrame:\n", df)

dfTfList = tf.convert_to_tensor(df)
print("Tensor DataFrame:\n", dfTfList)

print("--------CRUD--------")
print("Update and Delete operations are forbidden for tfList, due to immutability of tensor created by the .constant method")

tfListVar = tf.Variable([1,2,3,4,5])
print("Changeable tensor:\n", tfListVar)

tfListVar.assign([6,7,8,9,10])
print("Changed tensor:\n", tfListVar)

```
